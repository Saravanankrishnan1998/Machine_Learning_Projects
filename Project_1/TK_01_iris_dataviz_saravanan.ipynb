```python
# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./data/iris.csv')
```

```python
df.head()
```

```python
# Scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='SepalLengthCm', y='SepalWidthCm', hue='Species')
plt.title('Scatter Plot of Sepal Length vs Sepal Width')
plt.show()
```

```python
# Box plot
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Species', y='SepalLengthCm')
plt.title('Box Plot of Sepal Length by Species')
plt.show()
```

```python
# Histogram
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='SepalLengthCm', hue='Species', kde=True, element='step')
plt.title('Histogram of Sepal Length')
plt.show()
```

```python
# Pair plot
sns.pairplot(df, hue='Species')
plt.title('Pair Plot of Iris Dataset')
plt.show()
```

```python
# Violin plot
plt.figure(figsize=(10, 6))
sns.violinplot(data=df, x='Species', y='SepalWidthCm')
plt.title('Violin Plot of Sepal Width by Species')
plt.show()
```
